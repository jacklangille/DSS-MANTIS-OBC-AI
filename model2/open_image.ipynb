```python
#Save a .tiff
img1 = rasterio.open('2020-02-21.tiff')
```

```python
#Plot the RGB-NIR image
show(img1.read())
```

Saving raster image to a 3D-numpy array to perform pre-proccessign steps and inference.

```python
#set dimension of 3d-array to image height, width and number of bands
imgArray = np.zeros((img1.height, img1.width, img1.count))

#fill the array with rgbnir values
for i in range(4):
    imgArray[ :, :,i] = img1.read(i+1)


```
